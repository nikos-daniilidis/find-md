## Picking a doctor!

After the two previous posts discussing the collection and cleaning up of doctor ratings data, it is time for the fun part: Picking the best doctor. The poor man's approach is: Given the data, average all the ratings for each doctor, and whoever has the largest average rating wins. But the question we really want to answer is: Given the data, which doctor has the *highest probability* of having a high score, say above 4.8 out of 5?

More concretely, consider the following example: 

- Doctor A: 4 out of 4 ratings. Average rating: 4.5/5. Variance of ratings: 0.5/5.
- Doctor B: 4 out of 4 ratings. Average rating: 4.6/5. Variance of ratings: 0.3/5.
- Doctor C: 3 out of 4 ratings. Average rating: 4.5/5. Variance of ratings: 0.5/5.

It is not clear that one doctor is better than the others: the mean rating and standard deviation do not capture all of the information, and missing values complicate things even more. What we need is a *single measure* of quality for each doctor. We also want to be able to estimate the *probability distributions* of the doctors' quality measures, given the entire ensemble of ratings that we have in our possession. In what follows, I am using Principal Components Analysis (PCA) to derive a unique rating parameter from the data set, and boostraping to estimate the distribution of ratings for each doctor. 

Let me try to break this down, in case it was too technical: You can think of PCA as a way to draw a 'best fit' line (or plane in higher dimensions) through a cloud of data, by treating all the variables on equal footing (as opposed to least squares linear regression, which singles-out one of the vartiables to find the best fit). PCA allows me to compact my data from a $P$-dimensional space to lower dimensions, by picking those combinations of parameters which explain the maximal amount of variance in the data. In plain English: with PCA I can find a single linear combination of the four ratings (not necessarily the mean) which explains as much of the variance in the data as can be explained using a single parameter.  Moreover, [under certain assumptions](http://en.wikipedia.org/wiki/Principal_component_analysis#PCA_and_information_theory), the transformation is optimal in terms of compacting the information in the data to as few parameters as possible: If I keep the first component, I am guaranteed to have the maximum amount of information that I could have in a single parameter, given my observations.

So, here is the strategy: I will perform PCA on the data and keep only the component which explains the maximum amount of variance in the data. Let's call this component the 'PCA score'. PCA is not suited to data with missing values, so I will use one of the data imputation methods I developed before. I do not want to throw  away data, but also do not want to introduce too much bias. So, I will use stochastic mean imputation, and repeat the analysis many times to acquire good statistics. The randomness I introduce this way, will influence all PCA scores in a way which reflects the missing information in the data. This way, the PCA score of each doctor will become a probability distribution, based on which I can then select the best doctors. 

There are two more caveats here: First, by introducing normally distributed random ratings to impute the missing values, I am making an implicit assumption about the probability ditributions of missing ratings. Second, the algorithm described above, will be biased toward doctors who have all four ratings and have high scores in all four. For example, doctors with equally high, or higher ratings in only three websites will be less favored by the stochastic imputation step. To circumvent this, before the imputation I will add a step of randomly removing one out of four ratings from all doctors who have all four ratings. In summary, here is the algorithm I will use:

```
get rating data
repeat N times
{
    randmoly remove 1 out of 4 ratings where all 4 ratings are present
    do mean stochastic imputation on data
    do PCA on imputed data
    save PCA score for all doctors for this iteration
}
analyze the PCA score distributions from the N saved iterations
```

At the end of this process, I will have a *distribution* of PCA scores for each doctor. Then, I can use several different criteria to rate the doctors based  on the PCA score distributions.

Before I use PCA I will check, once more, that all of my parameters contain useful correlations to decide whether should keep all four ratings in my model. The best way to do this is by looking at the covariance matrix of my data. I will use the covariance matrix that comes with NumPy. The elements of the normalized covariance matrix are the maximum likelihood estimators of the covariance of my parameters. 

In what follows, I show the correlation coefficient matrices for the doctor rating data after aggresssive imputation, and after stochastic mean imputation. I plot the matrices on a blue colormap scale. Darker blue corresponds to high correlation. For display purposes, I have set the elements along the diagonal to zero. The values of those elements are normally one (everything is perfectly correlated with itself, except if it is zero or infinity). 

The qualitative picture is the same as the picture that I get when I use multiple linear regression to analyze the data. We see the same hierarchy of correlations betweeen ratings sources, and we also see that stochastic mean imputation results in spreading out of correlations over all parameters (let's call it "correlation diffusion" -it sounds sexy).

##### Repeat 1000 Imputation/PCA cycles. Double bootstrap

t is quite evident that any decision I make based on these distributions, will be biased in favor of doctors who received all four ratings. However, I find it hard to believe that doctors who got rated in all four websites are really that better than those who got, say, ratings in only three websites. 

There is a way around it, and it's a common trick: Throw out some data to get rid of some bias. Given that I do not want to be biased against doctors with less than four ratings, I will run the same algorithm, but I will only run it on data where I have randomly removed one out of four reviews of those doctors who have all four reviews. So here is the new algorithm:

```
get rating data
repeat N times
{
    randomly remove one out of four ratings wherever there are four ratings
    do mean stochastic imputation on data
    do PCA on imputed data
    save rating hyper-parameter for all doctors
}
```

### PCA Algorithms

````python
def pca_fit_transform_df(df,colsList,pca=decomposition.PCA(n_components=1),standardize=True,rangeout=1):
    '''Fit a PCA model to selected columns of a data frame.
    Returns a list with the reduced dimensionality features, and the 
    value of the explained variance ratio.
    '''
    X = df.as_matrix(columns=colsList)
    if standardize: # ATTENTION: this assumes integer scales with numbers close to max for all data!
        for i in range(np.shape(X)[1]):
            X[:,i] = X[:,i]/np.max(X[:,i]) 
    Xreduced = pca.fit_transform(X)
    # standardize to range 0-rangeout
    if standardize:
        mn = np.min(Xreduced)
        mx = np.max(Xreduced)
        Xreduced = rangeout*(Xreduced - mn)/(mx-mn)
    # align to the mean (slope is negative about half of the time)
    means= np.sum(df.as_matrix(columns=colsList),axis=1)/len(colsList)
    p = np.polyfit(means,Xreduced,1)
    if p[0]<0:
        Xreduced = 1.-Xreduced
    return Xreduced, pca.explained_variance_ratio_

def get_pca_score(df,colsList,Nrepetitions = 100,pca=decomposition.PCA(n_components=1), fullResults=True, dropRatings = False):
    '''Repeat Nrepetitions iterations of stochastic mean imputation and PCA transformation 
    on selected columns of the dataframe df. Return the mean and variance of the 
    PCA grade for each row of the datagrame df. 
    ---Inputs---
    df: data frame with the values before imputation, but with rows containing all NA removed
    colsList: columns of which the values I am interested in.
    Nrepetitions: number of repetitions (integer)
    pca: instantiation of the PCA decompostition object
    fullResults: Return the PCA score for all iteration (True), or return only mean and standard 
                deviation (False).
    dropRatings: if True, then randomly get rid of one out of the ratings where all of them are present
    ---Outputs---
    pcaScores: a numpy array with the PCA scores (fullResults==True)
    explainedVars: a numpy array with the explained variances (fullResults==True)
    
    meansCores: a numpy array with the mean PCA score for each doctor (fullResults==False)
    explainedVars: a numpy array with the standard deviation of the PCA score for each doctor (fullResults==False)
    '''
    pcaScores = np.zeros([df.shape[0],Nrepetitions])
    explainedVars = np.zeros([Nrepetitions,1])
    meanScores = np.zeros([pcaScores.shape[0],1])
    stdevScores = np.zeros([pcaScores.shape[0],1])
    for i in range(Nrepetitions):
        msi_df = mean_xxpute_df(df,colsList,stochastic=True,min_sd=0.1,drop=dropRatings)
        grd, vr = pca_fit_transform_df(msi_df,colsList,pca)
        pcaScores[:,i] = grd.T
        explainedVars[i] = vr
    if fullResults:
        return pcaScores, explainedVars    
    print 'Mean explained variance: %3.2f'%np.mean(explainedVaars)
    for j in range(pcaScores.shape[0]):
        meanScores[j] = np.mean(pcaScores[j,:])
        stdevScores[j] = np.std(pcaScores[j,:])
    return means,stdevs

def old_pca_score(df,colsList,Nrepetitions = 100,pca=decomposition.PCA(n_components=1), fullResults=True, dropRatings = False):
    '''Repeat Nrepetitions iterations of stochastic mean imputation and PCA transformation 
    on selected columns of the dataframe df. Return the mean and variance of the 
    PCA grade for each row of the datagrame df. 
    ---Inputs---
    df: data frame with the values before imputation, but with rows containing all NA removed
    colsList: columns of which the values I am interested in.
    Nrepetitions: number of repetitions (integer)
    pca: instantiation of the PCA decompostition object
    fullResults: Return the PCA score for all iteration (True), or return only mean and standard 
                deviation (False).
    ---Outputs---
    pcaScores: a numpy array with the PCA scores (fullResults==True)
    explainedVars: a numpy array with the explained variances (fullResults==True)
    
    meansCores: a numpy array with the mean PCA score for each doctor (fullResults==False)
    explainedVars: a numpy array with the standard deviation of the PCA score for each doctor (fullResults==False)
    '''
    pcaScores = np.zeros([df.shape[0],Nrepetitions])
    explainedVars = np.zeros([Nrepetitions,1])
    meanScores = np.zeros([pcaScores.shape[0],1])
    stdevScores = np.zeros([pcaScores.shape[0],1])
    for i in range(Nrepetitions):
        msi_df = mean_impute_df(df,colsList,stochastic=True)
        grd, vr = pca_fit_transform_df(msi_df,colsList,pca)
        pcaScores[:,i] = grd.T
        explainedVars[i] = vr
    if fullResults:
        return pcaScores, explainedVars    
    print 'Mean explained variance: %3.2f'%np.mean(explainedVaars)
    for j in range(pcaScores.shape[0]):
        meanScores[j] = np.mean(pcaScores[j,:])
        stdevScores[j] = np.std(pcaScores[j,:])
    return means,stdevs
````

### Post-PCA visualization and processsing

````python
def score_histogram(scoresMatrix, rowNum, titleStr='', 
                    semilog=False, semilogEps=0.001, 
                    bins=60, normed=1, yScale=0.6, rang=[0.,1.]):
    '''Plot histogram of values in rowNum row of scoreMatrix. 
    If semilog, the values are plotted on a semilog x scale.
    ---Inputs---
    scoresMatrix: numy arrray. I will make histogram along one row of scoresMatrix
    rowNum: the row of which to make the historgram
    titleStr: title of the plot
    '''
    fig = plt.figure()
    if semilog:
        p = plt.hist(np.log(1+semilogEps-scoresMatrix[rowNum,:]),bins=60, normed=1) 
    else:
        p = plt.hist(scoresMatrix[rowNum,:],bins=bins, range=rang, normed=normed)
    ax = fig.gca()
    ax.set_title(titleStr)
    if semilog:
        ax.set_xlabel('log(1+eps-PCA score)',fontsize=15)
    else:
        ax.set_xlim(rang[0],rang[1])
        ax.set_ylim(0.,round(bins*yScale))
        ax.set_xlabel('PCA score',fontsize=15)
        ax.set_ylabel('normalized occurrences',fontsize=15)
    ax.set_ylabel('normalized occurrences',fontsize=15)   

def score_histogram_matrix(rowsList, titlesList,panelDims = (2,2), figsize=[6,4],
                        bins=60, normed=1, yScale=0.6, rang=[0.,1.]):
    '''Plot a panel of histograms for a list of numpy arrays.
    ---Inputs---
    rowsList: a list of numpy arrays
    titlesList: list of strings, one title per array
    panelDims: list or tuple, specifying how to organize the histograms. 
               Default is 2 by 2
    '''
    fig, axes=plt.subplots(figsize=figsize, nrows=panelDims[0], ncols=panelDims[1])
    roCount = 0
    for i in range(panelDims[0]):
        for j in range(panelDims[1]):
            axes[i][j].hist(rowsList[roCount], bins=bins, range=rang, normed=normed)
            axes[i][j].set_title(titlesList[roCount],fontsize=8)
            axes[i][j].set_xlabel('PCA score',fontsize=8) 
            axes[i][j].set_ylabel('normalized occurrences',fontsize=8) 
            axes[i][j].tick_params(axis='both', which='major', labelsize=8)
            axes[i][j].set_xlim(rang[0],rang[1])
            axes[i][j].set_ylim(0.,round(bins*yScale))
            roCount += 1
    fig.tight_layout()
  
````

````python
def above_threshold_probabilities_df(scoreDf,bootStrapCols,thresholdList):
    '''Make a dataframe with the probability for each doctor to have a PCA
    score above a certain threshold
    ---Inputs---
    scoreDf: dataframe with the PCA score results from the PCA bootstrap
            analysis
    bootstrapCols: a list with the names of the columns from scoreDf that I 
                want to use to estimate thresholds
    thresholdList: a list of score thresholds fo which the probabilities will 
                be calculated
    ---Outputs---
    df: a dataframe with the doctor names and several columns (one for each 
        threshold) with the estimated probabilities for each doctor's score 
        to exceed the given threshold value.
    '''
    df = pd.DataFrame(data=scoreDf['name'],index=scoreDf.index, columns=['name'])
    for thresh in thresholdList:
        col = np.zeros([df.shape[0],1])
        for ix in scoreDf.index:
            col[ix,0] = sum([float(x>thresh)for x in scoreDf.loc[ix,bootStrapCols]])/len(bootStrapCols)
        newName = 'P(score>%6.2f)'%thresh
        df[newName]=col
    return df
````

````python
def clean_series_data(x_ser, y_ser):
    '''Return numpy arrays with the data in x_ser, y_ser, with missing values removed'''
    x = np.array(x_ser.tolist())
    y = np.array(y_ser.tolist())
    missingx = np.isnan(x)
    x = x[~missingx]
    y = y[~missingx]
    missingy = np.isnan(y)
    x = x[~missingy]
    y = y[~missingy]
    return np.array(x),np.array(y)
    
def column_clean_polyfit(x_ser,y_ser,order,ycutoff = 0):
    '''Performs a polynomial fit on pandas series data. Gets rid of missing values.
    ycutoff: lower cutoff for filtering the y values
    Returns the fit coefficients'''
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    fitparams = np.polyfit(x,y,order)
    return fitparams

def plot_linfit_conf(x_ser,y_ser,p,confint,ax,ycutoff = 0., fontsize=8, *kwargs):
    '''Plot confidence intervals on the linear fit for the x_ser, y_ser data
    represented by p. 
    x_ser, y_ser: pandas series with x and y data
    p: polynomial fit coefficients returned from polyfit
    confint: string specifying the confidence interval to be used. Allowed values: 
            '50' '60' '70' '80' '90' '95' '96' '98' '99' '99.5' '99.8' '99.9'
    ax: an axis object for the plot
    ycutoff: lower cutoff for filtering the y values'''
    
    # select confidence interval to plot
    tdict = {'50':0.674, '60':0.841, '70':1.036, '80':1.282, '90':1.645, '95':1.960,
             '96':2.054, '98':2.326, '99':2.576, '99.5':2.807, '99.8':3.091, '99.9':3.291}
    try:
        t = tdict[confint]
    except: 
        t = tdict['70']
    
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    # find statistics
    N = len(x)
    Y = np.polyval(p,x)
    sigmasq = np.sum(np.power(y-Y,2))/(N-2)
    xmean  = np.mean(x)
    ymean = np.mean(y)
    SSt = np.sum(np.power(y-ymean,2))
    SSr = np.sum(np.power(Y-ymean,2))
    Rsq = SSr/SSt
    Sxx = np.sum(np.power(x-xmean,2))
    xp = np.linspace(0.9*np.min(x),1.1*np.max(x))
    yp = np.polyval(p,xp)
    delta = t*np.sqrt( sigmasq* ( 1./N + np.power(xp-xmean,2)/(Sxx) ) )
    yplus = yp+delta
    yminus = yp-delta
    ttl = r'$R^{2}$= %4.3f'%Rsq
    
    # plot
    ax.plot(xp,yp, color = 'black', alpha = 0.4)
    ax.plot(xp,yplus,'black', alpha = 0.2)
    ax.plot(xp,yminus,'black', alpha = 0.2)
    ax.fill_between(xp,yminus,yplus,color = 'grey', alpha = 0.3)
    ax.set_title(ttl, fontsize=fontsize)
````

### Mean imputation

````python
def in_range(r,low,high):
    '''If low<=r<=high, return r. Else, return low if r<low and high if r>high'''
    return float(high<r)*high+float(r<low)*low+float(low<=r<=high)*r

def replace_by_mean(sr, stochastic = False, min_sd = 0.1, low=0., high=1.):
    '''Replace the missing values from a pandas series by the mean of the
    existing values. 
    
    The input values *must* be in the range [low,high], which defaults to [0,1].
    
    If stochastic is set to True, the imputed values for each row are drawn 
    from a normal distribution with mean the mean of the non-NaN values in 
    the row, and standard deviation the standard deviation of the non-NaN values 
    in that row. If the standard deviation is zero, the standard deviation is set to 
    min_sd of the range of allowed values (which defaults to 0.1).
    
    Returns a list.
    
    To do: pass pre-generated random numbers from the caller function.
    '''
    lst = sr.values.tolist()
    mn = np.mean([el for el in lst if not np.isnan(el)])
    sd = max([np.std([el for el in lst if not np.isnan(el)]),min_sd])
    if np.isnan(mn):
        print('Warning: Encountered all NaN values in replace_by_mean')
    new_lst = []
    for el in lst:
        if np.isnan(el):
            if stochastic:
                new_lst.append(in_range(np.random.normal(mn,sd),low,high))
            else:
                new_lst.append(mn)
        else:
            new_lst.append(el)
    return new_lst
#replace_by_mean(grades_df.iloc[0])
````

````python
def mean_impute_df(df_in,columns_list, stochastic = False, min_sd = 0.1):
    '''
    Impute the NaN numerical values among columns_list using the mean of 
    the non-NaN numerical values on the same row of df_in. 
    
    Removes the rows where all nuberical entries are NaN. 
    
    If stochastic is set to True, the imputed values for each row are drawn 
    from a normal distribution with mean the mean of the non-NaN values in 
    the row, and standard deviation the standard deviation of the non-NaN values 
    in that row.
    
    Returns df_out, a dataframe
    '''
    # get rid of rows with all NaN ratings. YaN stands for Yes a  Number
    df_out = df_in.dropna(how='all', subset=columns_list).reset_index()
    max_values = {}
    
    # renormalize all values to interval 0-1
    for col in columns_list:
        max_values[col]=round(df_out[col].max())
        df_out[col] /= max_values[col]
    
    # run through rows and replace NaN by the mean of that row, where necessary
    for idx in df_out.index:
        #print ix
        try:
            df_out.loc[idx,columns_list] = replace_by_mean(df_out.iloc[idx][columns_list],
                                                       stochastic,
                                                       min_sd,
                                                       low = 0.2,
                                                       high = 1.)
        except:
            # if something went wrong, print the entry
            print 'Encountered problem with mean imputation in:'
            print df_out.iloc[idx]
            
    # rescale the columns back to the original ranges
    for col in columns_list:
        df_out[col] *= max_values[col]
    
    return df_out
````

### Multiple Linear Regression on Dataframes

````python
def multiple_lin_reg_df(df,colsList,yIndex,confInt):
    '''Multiple linear regression for selected columns from a o pandas dataframe.
    Fits to the model y = beta0 + beta1*x1 + ... + betaK*xK.
    Returns regression coefficients and their confidence intervals. 
    
    ---Inputs--- 
    df: input dataframe
    colsList: the columns of the data frame on which to fit the model (list)
    yIndex: the index of the y (dependent) column in cols_lst (integer)
    confInt : string specifying the confidence interval to be used. Allowed values: 
              '50' '60' '70' '80' '90' '95' '96' '98' '99' '99.5' '99.8' '99.9'
              Defaults to '70'
    
    ---Outputs---
    beta: list of coefficients [b0, b1, ..., bK]. Intercept is first
    deltaBeta: list of [0,a/2) confidence intervlas for the coefficients. Same ordering as beta.
    deltaY: list of confidence intervlas for the observed y values in the sample
    C: variance covariance matrix
    Rsq: Adjusted R^2 statistic incl;uding the degreess of freedom in the model 
         Rsq = 1-(n-1)/(n-p).(1-R^2)
         
    Source: http://reliawiki.org/index.php/Multiple_Linear_Regression_Analysis#Estimating_Regression_Models_Using_Least_Squares
    '''
    # select confidence interval to plot
    tdict = {'50':0.674, '60':0.841, '70':1.036, '80':1.282, '90':1.645, '95':1.960,
             '96':2.054, '98':2.326, '99':2.576, '99.5':2.807, '99.8':3.091, '99.9':3.291}
    try:
        t = tdict[confInt]
    except: 
        t = tdict['70']
    
    # drop rows with any missing values
    sub_df = df[colsList].dropna(how='any')
    
    # I have marked the dependent column by yIndex. Select y and x values from df
    y = np.array(sub_df[colsList[yIndex]].tolist()) # N samples by 1
    X = sub_df.drop(colsList[yIndex],axis=1).as_matrix() # N samples by K slopes
    N = np.shape(y)[0] # N samples
    K = np.shape(X)[1] # P = K+1 features (K slopes + 1 intercept) 
    
    #standardize: standardize the scales of the data. Not necesssary for lin.reg.?
    # standardize the scales (subtract mean, )
    #Ymax = np.rint(np.max(yvals))
    #Xmax = np.rint(np.max(xvals,axis=0))
    #y = yvals/Ymax
    #X = xvals/Xmax
    
    #add ones to the beginning of X
    X = np.concatenate([np.ones([np.shape(X)[0],1]),X], axis=1)
    
    iXX = np.linalg.inv(np.dot(X.T,X))
    H = np.dot(X,np.dot(iXX,X.T))
    
    # estimates for regression coefficients, y values, 
    beta = np.dot(np.dot(iXX,X.T),y)
    Yest = np.dot(X,beta)
    
    # mean squared errors
    I = np.eye(N)
    J = np.ones([N,N])
    SSe = np.dot(y.T,np.dot(I-H,y))
    MSe = SSe/(N-K-1)
    SSt = np.dot(y.T,np.dot(I-(1/N)*J,y))
    MSt = SSt/(N-1)
    SSr = np.dot(y.T,np.dot(H-(1/N)*J,y))
    MSr = SSr/(N-1)
    #print SSe,MSe,SSt,MSt,SSr,MSr
    
    # estimate of the variance
    SigmaSq = MSe
    
    # Variance-Covariance matrix
    C = SigmaSq*iXX
    
    # Coefficient of multiple derermination
    Rp2 = 1-SSe/SSt
    # Adjusted for number of dof
    Rsq = 1-(N-1)/(N-K-1)*(1-Rp2)
    
    deltaBeta = t*np.array([np.sqrt(C[i,i]) for i in range(K+1)])
    deltaY = np.diag(t*np.power(SigmaSq*np.dot(X,np.dot(iXX,X.T)),0.5))
    
    return beta,deltaBeta,deltaY, C, Rsq
````

#### Use linear regression to map correlations between variables

````python
def regressionCorrelationMap(df,colsList,confInt='70',verbose=False):
    '''Return a map of correlations between all variables in selected ccolumns of a dataframe.
    Fit  linear regression models to all the columns by iterating the dependent variable through 
    colsList. Record the values of linear coefficients divided by their confidence intervals.
    ---Inputs---
    df: input dataframe
    colsList: the column names on which the multiple regression is fitted
    confInt: confidence interval level. Defaults to '70'
    ---Outputs---
    corrMap: a numpy aarray
    '''
    corrMap = np.zeros([len(colsList),len(colsList)])
    for  ind,col in enumerate(colsList):
        beta,deltaBeta,deltaY,C,Rsq = multiple_lin_reg_df(df,colsList,yIndex=ind,confInt=confInt)
        ls = range(4)
        ls.remove(ind)
        corrMap[ind,ls[0]] = beta[1]/deltaBeta[1]
        corrMap[ind,ls[1]] = beta[2]/deltaBeta[2]
        corrMap[ind,ls[2]] = beta[3]/deltaBeta[3]
        if verbose:
            print 'Y=%s'%col
            print 'Intercept: (%2.1e,%2.1e). Mean/ConfInt = %d'%(beta[0],deltaBeta[0],round(beta[0]/deltaBeta[0]))
            print 'X1 = %s, β1->(%2.1e,%2.1e). Mean/ConfInt = %d'%(cols_lst[ls[0]],beta[1],deltaBeta[1],round(beta[1]/deltaBeta[1]))
            print 'X2 = %s, β2->(%2.1e,%2.1e). Mean/ConfInt = %d'%(cols_lst[ls[1]],beta[2],deltaBeta[2],round(beta[2]/deltaBeta[2]))
            print 'X3 = %s, β3->(%2.1e,%2.1e). Mean/ConfInt = %d'%(cols_lst[ls[2]],beta[3],deltaBeta[3],round(beta[3]/deltaBeta[3]))
            print 'R^2=%3.2f\n'%Rsq
    return corrMap
````
