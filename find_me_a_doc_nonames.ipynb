They are either HMO plans or PPO plans. Roughly speaking, with an HMO plan you have lower co-payments and healthcare costs, but you are extremely restricted on the doctors you are allowed to see. Most of your health issues have to go through a specific doctor, your 'primary care' doctor. If you get a hurting foot, you have to go through your primary care doctor, who will send you to the foot doctor. With a PPO plan you have more freedom, you can see the foot doctor directly, as long as they are within a network of allowed providers. ##### Plotting functions

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
def clean_series_data(x_ser, y_ser):
    '''Return numpy arrays with the data in x_ser, y_ser, with missing values removed'''
    x = np.array(x_ser.tolist())
    y = np.array(y_ser.tolist())
    missingx = np.isnan(x)
    x = x[~missingx]
    y = y[~missingx]
    missingy = np.isnan(y)
    x = x[~missingy]
    y = y[~missingy]
    return np.array(x),np.array(y)
    
def column_clean_polyfit(x_ser,y_ser,order,ycutoff = 0):
    '''Performs a polynomial fit on pandas series data. Gets rid of missing values.
    ycutoff: lower cutoff for filtering the y values
    Returns the fit coefficients'''
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    fitparams = np.polyfit(x,y,order)
    return fitparams

def plot_linfit_conf(x_ser,y_ser,p,cnfint,ax,ycutoff = 0.,*kwargs):
    '''Plot confidence band on the linear fit for the x_ser, y_ser data
    represented by p. The title of the plot is the R^2 coefficient for the 
    linear fit to the data. 
    x_ser, y_ser: pandas series with x and y data
    p: polynomial fit coefficients returned from polyfit
    cnfint: string specifying the confidence interval to be used. Allowed values: 
            '50' '60' '70' '80' '90' '95' '96' '98' '99' '99.5' '99.8' '99.9'
    ax: an axis object for the plot
    ycutoff: lower cutoff for filtering the y values'''
    
    # select confidence interval to plot
    tdict = {'50':0.674, '60':0.841, '70':1.036, '80':1.282, '90':1.645, '95':1.960,
             '96':2.054, '98':2.326, '99':2.576, '99.5':2.807, '99.8':3.091, '99.9':3.291}
    try:
        t = tdict[cnfint]
    except: 
        t = tdict['70']
    
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    # find statistics
    N = len(x)
    Y = np.polyval(p,x)
    sigmasq = np.sum(np.power(y-Y,2))/(N-2)
    xmean  = np.mean(x)
    ymean = np.mean(y)
    SSt = np.sum(np.power(y-ymean,2))
    SSr = np.sum(np.power(Y-ymean,2))
    Rsq = SSr/SSt
    Sxx = np.sum(np.power(x-xmean,2))
    xp = np.linspace(0.9*np.min(x),1.1*np.max(x))
    yp = np.polyval(p,xp)
    delta = t*np.sqrt( sigmasq* ( 1./N + np.power(xp-xmean,2)/(Sxx) ) )
    yplus = yp+delta
    yminus = yp-delta
    ttl = r'$R^{2}$= %4.3f'%Rsq
    
    # plot
    ax.plot(xp,yp, color = 'black', alpha = 0.4)
    ax.plot(xp,yplus,'black', alpha = 0.2)
    ax.plot(xp,yminus,'black', alpha = 0.2)
    ax.fill_between(xp,yminus,yplus,color = 'grey', alpha = 0.3)
    ax.set_title(ttl)
    return 

# some example data
xx = [3.,4.,5.,6.,7.,8.]
yy = [0.9,2.1,2.9,4.1,4.9,6.1]
X = pd.Series(xx)
Y = pd.Series(yy)
p = column_clean_polyfit(X,Y,1)
fig = plt.figure()
ax = fig.gca()
ax.scatter(X,Y)
plot_linfit_conf(X,Y,p,'95',ax)
plt.show()
```

##### Make scatter plots and histograms

```python
websites=['healthgrade','ratemds','ucomphealthcare','vitals']

fig, axes=plt.subplots(figsize=(12,12), nrows=len(websites), ncols=len(websites))
for i, websitex in enumerate(websites):
    for j, websitey in enumerate(websites):
        if i==j:
            x = np.array(doc_rating_df[websitex+'_rating'].dropna().tolist())
            axes[i][j].hist(x, bins=[0.,1.,2.,3.,4.,5.,6.], normed=True, color='b', alpha = 0.3)
            axes[i][j].set_xlabel(websitex+'.com rating ') 
            axes[i][j].set_ylabel('P(rating)') 
            axes[i][j].axis([0, 6, 0, 1])
        else:
            X = doc_rating_df[websitex+'_rating']
            Y = doc_rating_df[websitey+'_rating']
            p = column_clean_polyfit(X,Y,1)
            x,y = clean_series_data(X,Y)
            plot_linfit_conf(X,Y,p,'95',axes[i][j])
            axes[i][j].scatter(x,y, alpha = 0.2)
            axes[i][j].set_xlabel(websitex+'.com rating ')
            axes[i][j].set_ylabel(websitey+'.com rating ')
            axes[i][j].axis([0, 6, 0, 6])
fig.tight_layout()
```

##### Discussion 

Well, what have we here? The only thing that's clear is that the data is all over the place! But first, what are we looking at?
     
Along the diagonal, we are  looking at the histogram of ratings for each of the four websites, i.e. the histogram for healthgrade.com, the histogram for ratemds.com, the histogram for ucomphealthcare.com, and the one for vitals.com. The maximum rating for the first three is five stars, while vitals.com goes only to four stars. Note that even the shape of the histograms is different, i.e. different websites show different distributions (for my small sample). Three out of four seem to have inflated ratings, and only ratemds.com seems to have the bell-shaped form I would expect. Why do I expect a bell shape? Because it is more likely that some doctors will be excellent and some will be bad, but most doctors will be quite good. If all doctors are outstanding, I either have the problem that my doctor pool has been truncated to leave out the better ones (so that I am sampling the lower half of a bell curve), or I have the problem of grade inflation, which somehow followed all these doctors after they were out of medical school. Which of the two it is, I do not know.
    
However, the correlation of ratings between different websites is where the world falls apart! This is what you see in the curves off the diagonal, for exammple the second plot in the first row shows the ratings in ratemds.com vesrus the ratings in healthgrade.com. Each point in these off-diagonal plots is a doctor. Also alll points are identicaland semitransparent, which means that darker blue dots reveal a high concentration of doctors in a particular region. If there was good correlation for the grades of the sane doctor in different websites, I would expect the points to fall on a line with some scatter. That's not what I see! To see how bad this is, I try to fit a line through the data. This is the bold black line that you see in each of the curves. Around each line, I draw, as a grey band, the 95% confidence interval for my linear fit. This band represents the range of values where 95% of my data should fall, if the linear fit is a good description of my data. Are the linear fits good descriptions? No! You can get an idea of how bad they are, by looking at the $R^2$ value which is the title of each plot. You can take this value as giving you the fraction of variability in the data which is described by the linear model. Some pairs of websites give extremely bad results, with $R^2$ as low as 0.004 for the ratings between ucomphealthcare.com and healthgrade.com. The biggest degree of linear correlation, around 0.25, is between ucomphealthcare.com and vitals.com. 

To finish on a positive note, not all is bleak. A closer look reveals that while the low-rated doctors seem to show a lot of scatter, high-rated doctors tend to be highly rated in more than one website. For example, there are concentrations of dark points at the high grade regions in vitals.com, ratemds.com, and ucomphealthcare.com. This allows me to pick a good doctor with relatively high confidence. I have to pick someone who rates well in all of the available websites. This is not the most mathematically proper decision procedure, but it allows me to at least choose a primary care physician, before I can fully develop the proper decision algorithm.

I might follow up post on this with some more work on making sense out of this data. One idea to make use of all the ratings, is to do what is called principal components analysis (PCA) on the data. Principal components analysis is a standard trick for situations where you have (possibly noisy) data in a high-dimensional space, and you want to find those combinations of dimensions which contain most of the information which is hidden in your data. Then you can keep a few parameters (which are linear combinations of all dimensions) which contain most of the information in your data, and not worry about the rest. In other words PCA would allow me to generate a "hypergrade", one single parameter which tells me who is the best doctor, givem the data at hand.
