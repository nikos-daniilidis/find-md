## Looking at doctor ratings with data imputation

I am having a second look at doctor ratings in the SF bay area. This time, I collected more data, and I implemented a few additional methods, mostly to deal with missing data. 

I used two methods of data imputation (filling in missing data) for the doctor ratings. The crude method was: just drop all doctors who have any missing ratings. Obviously, this ends up in a massive loss of data, and can add some bias. The second method I tried was mean imputation: for each doctor fill-in the missing ratings by using the mean of the existing ratings. This is still fairly crude and could also add some bias. A slight improvement is possible by adding a stochastic element: whenever there are missing values, generate the missing values as random draws from a normal distribution with the mean and the standard deviation of the existing values. I have implemented these two methods in functions which work on Pandas dataframes, although the former method (let's call it aggresssive imputation) is a one-liner in Pandas. 

The second thing I implemented was multiple linear regression on Pandas dataframes. I wrote a little function which does the regression and returns confidence intervals and $R^2$ values. This offers a handy way to evaluate correlations among ratings. It also allows evaluating how much of the scatter in the plots of my [last blog post](http://oligotropos.wordpress.com/2014/06/25/staying-healthy-staying-sane/) can be explained once I include more variables to my model.

I implemented two methods of mean imputation. "Brute" mean imputation replaces missing values with the means of existing corresponding values. "Stochastic" mean imputation replaces the missing values with random draws from a normal distribution with the mean and standard deviation of the existing corresponding values. Here I have a look at how they perform:

##### ---Stochastic mean imputation---

The large density of points on the diagonals of the scatter plots reveals a limitation of the "brute" mean imputation method. Here's why: If I am missing two out of four ratings for a particular doctor, then I will fill their values with the same number (the mean of the existing ones). When I make a scatter-plot, these values will appear on the diagonal, creating an illusion of perfect correlation - a kind of false optimism of the method. This is why stochastic mean imputation is a good idea: By adding a random element to the imputed values, I reduce this fictitious correlation and counter balance the false optimism. In terms of fitting models to the imputed data, we can expect that adding the stochastic element will reduce  the amount of bias we introduce because  of the fictitious correlations. 

Here  is how  this looks. We now replace  the missing values with random draws from a normal distribution with the same mean and variance as those of the existing values.

```python
cols_lst = ['healthgrade_rating', 'ratemds_rating', 'ucomphealthcare_rating', 'vitals_rating']
MeanImputStoch_df = mean_impute_df(doc_rating_df,cols_lst, stochastic=True)
MeanImputStoch_df.head()
```

```python
scatter_matrix(MeanImputStoch_df,websites)
```

Using stochastic mean imputation, the values are now more distributed in the scatter plots. Note that I have instructed the imputation algorithm to limit the values of the imputed results to within the valid range of each website. This  is a necessary step if you do not want to end up with negative ratings, or ratings of 5.3  on a scale of 1-5. 

As you can see, I am now getting a lower degree of correlation than I was before. This is clear from the $R^2$ values in the linear fits, which are now 10-30% lower than they were after the "brute" mean imputation. This is a  good thing! It means we have partially solved the problem of over-optimism of our model. 

One last thing to note, is that the linear regression model I am using to fit a line through the scatterplots suffers from "regression toward the mean". In other words, linear regression tends to give predictions which are closer to the mean than to the extremes. It is a conservative model! The  most obvious symptom of this is when your reaction to looking at a fit is: "this fit is wrong, it is way too flat". Looking at the scatter plots in both data sets with mean imputation, the linear fits certainly have this look. It is not a problem of the algorithm, but rather a feature  of the linear regression model that you are looking at.

### Mean imputation

```python
def in_range(r,low,high):
    '''If low<=r<=high, return r. Else, return low if r<low and high if r>high'''
    return float(high<r)*high+float(r<low)*low+float(low<=r<=high)*r

def replace_by_mean(sr, stochastic = False, min_sd = 0.1, low=0., high=1.):
    '''Replace the missing values from a pandas series by the mean of the
    existing values. 
    
    The input values *must* be in the range [low,high], which defaults to [0,1].
    
    If stochastic is set to True, the imputed values for each row are drawn 
    from a normal distribution with mean the mean of the non-NaN values in 
    the row, and standard deviation the standard deviation of the non-NaN values 
    in that row. If the standard deviation is zero, the standard deviation is set to 
    min_sd of the range of allowed values (which defaults to 0.1).
    
    Returns a list.
    
    To do: pass pre-generated random numbers from the caller function.
    '''
    lst = sr.values.tolist()
    mn = np.mean([el for el in lst if not np.isnan(el)])
    sd = max([np.std([el for el in lst if not np.isnan(el)]),min_sd])
    if np.isnan(mn):
        print('Warning: Encountered all NaN values in replace_by_mean')
    new_lst = []
    for el in lst:
        if np.isnan(el):
            if stochastic:
                new_lst.append(in_range(np.random.normal(mn,sd),low,high))
            else:
                new_lst.append(mn)
        else:
            new_lst.append(el)
    return new_lst
#replace_by_mean(grades_df.iloc[0])
```

```python
def mean_impute_df(df_in,columns_list, stochastic = False, min_sd = 0.1):
    '''
    Impute the NaN numerical values among columns_list using the mean of 
    the non-NaN numerical values on the same row of df_in. 
    
    Removes the rows where all nuberical entries are NaN. 
    
    If stochastic is set to True, the imputed values for each row are drawn 
    from a normal distribution with mean the mean of the non-NaN values in 
    the row, and standard deviation the standard deviation of the non-NaN values 
    in that row.
    
    Returns df_out, a dataframe
    '''
    # get rid of rows with all NaN ratings. YaN stands for Yes a  Number
    df_out = df_in.dropna(how='all', subset=columns_list).reset_index()
    max_values = {}
    
    # renormalize all values to interval 0-1
    for col in columns_list:
        max_values[col]=round(df_out[col].max())
        df_out[col] /= max_values[col]
    
    # run through rows and replace NaN by the mean of that row, where necessary
    for idx in df_out.index:
        #print ix
        try:
            df_out.loc[idx,columns_list] = replace_by_mean(df_out.iloc[idx][columns_list],
                                                       stochastic,
                                                       min_sd,
                                                       low = 0.2,
                                                       high = 1.)
        except:
            # if something went wrong, print the entry
            print 'Encountered problem with mean imputation in:'
            print df_out.iloc[idx]
            
    # rescale the columns back to the original ranges
    for col in columns_list:
        df_out[col] *= max_values[col]
    
    return df_out
```

```python
def imshow_array(arr, title, xlabels, ylabels, interpolation='none', cmap='Blues'):
    '''
    Plot the contents of a numpy array as a colorscale image. Manually set the title and tick labels
    '''
    fig = plt.figure()
    plt.imshow(arr, interpolation='none', cmap='Blues')
    plt.xlabel('website x',fontsize=15)
    plt.ylabel('website y',fontsize=15)
    ax = fig.gca()
    ax.set_title(title,fontsize=15)
    ax.set_xticks(range(np.shape(arr)[0]))
    ax.set_yticks(range(np.shape(arr)[0]))
    ax.set_xticklabels(xlabels)
    ax.set_yticklabels(ylabels)
    plt.show()
```

```python
def scatter_matrix(df,colnames,fontsize=8):
    '''Create scatter plots of the numbers in the columns of a dataframe
    Along the diagonal make histograms. 
    df is a dataframe which contains the columns in colnames. The values 
    in colnames are numeric.
    '''
    fig, axes=plt.subplots(figsize=(9,8), nrows=len(colnames), ncols=len(colnames))
    for i, colx in enumerate(colnames):
        for j, coly in enumerate(colnames):
            if i==j:
                x = np.array(df[colx+'_rating'].dropna().tolist())
                axes[i][j].hist(x, bins=[0.,1.,2.,3.,4.,5.,6.], normed=True, color='b', alpha = 0.3)
                axes[i][j].set_xlabel(colx+'.com rating ', fontsize=fontsize) 
                axes[i][j].set_ylabel('P(rating)', fontsize=fontsize) 
                axes[i][j].axis([0, 6, 0, 1])
            else:
                X = df[colx+'_rating']
                Y = df[coly+'_rating']
                p = mpa.column_clean_polyfit(X,Y,1)
                mpa.plot_linfit_conf(X,Y,p,'95',axes[i][j], fontsize=fontsize)
                axes[i][j].scatter(X,Y, alpha = 0.2)
                axes[i][j].set_xlabel(colx+'.com rating ', fontsize=fontsize)
                axes[i][j].set_ylabel(coly+'.com rating ', fontsize=fontsize)
                axes[i][j].axis([0, 6, 0, 6])
    fig.tight_layout()
    plt.show()
```

```python
def clean_series_data(x_ser, y_ser):
    '''Return numpy arrays with the data in x_ser, y_ser, with missing values removed'''
    x = np.array(x_ser.tolist())
    y = np.array(y_ser.tolist())
    missingx = np.isnan(x)
    x = x[~missingx]
    y = y[~missingx]
    missingy = np.isnan(y)
    x = x[~missingy]
    y = y[~missingy]
    return np.array(x),np.array(y)
    
def column_clean_polyfit(x_ser,y_ser,order,ycutoff = 0):
    '''Performs a polynomial fit on pandas series data. Gets rid of missing values.
    ycutoff: lower cutoff for filtering the y values
    Returns the fit coefficients'''
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    fitparams = np.polyfit(x,y,order)
    return fitparams

def plot_linfit_conf(x_ser,y_ser,p,confint,ax,ycutoff = 0., fontsize=8, *kwargs):
    '''Plot confidence intervals on the linear fit for the x_ser, y_ser data
    represented by p. 
    x_ser, y_ser: pandas series with x and y data
    p: polynomial fit coefficients returned from polyfit
    confint: string specifying the confidence interval to be used. Allowed values: 
            '50' '60' '70' '80' '90' '95' '96' '98' '99' '99.5' '99.8' '99.9'
    ax: an axis object for the plot
    ycutoff: lower cutoff for filtering the y values'''
    
    # select confidence interval to plot
    tdict = {'50':0.674, '60':0.841, '70':1.036, '80':1.282, '90':1.645, '95':1.960,
             '96':2.054, '98':2.326, '99':2.576, '99.5':2.807, '99.8':3.091, '99.9':3.291}
    try:
        t = tdict[confint]
    except: 
        t = tdict['70']
    
    # clean data
    Yc = ycutoff*y_ser.max()
    x_ser = x_ser[y_ser>Yc]
    y_ser = y_ser[y_ser>Yc]
    x,y = clean_series_data(x_ser,y_ser)
    
    # find statistics
    N = len(x)
    Y = np.polyval(p,x)
    sigmasq = np.sum(np.power(y-Y,2))/(N-2)
    xmean  = np.mean(x)
    ymean = np.mean(y)
    SSt = np.sum(np.power(y-ymean,2))
    SSr = np.sum(np.power(Y-ymean,2))
    Rsq = SSr/SSt
    Sxx = np.sum(np.power(x-xmean,2))
    xp = np.linspace(0.9*np.min(x),1.1*np.max(x))
    yp = np.polyval(p,xp)
    delta = t*np.sqrt( sigmasq* ( 1./N + np.power(xp-xmean,2)/(Sxx) ) )
    yplus = yp+delta
    yminus = yp-delta
    ttl = r'$R^{2}$= %4.3f'%Rsq
    
    # plot
    ax.plot(xp,yp, color = 'black', alpha = 0.4)
    ax.plot(xp,yplus,'black', alpha = 0.2)
    ax.plot(xp,yminus,'black', alpha = 0.2)
    ax.fill_between(xp,yminus,yplus,color = 'grey', alpha = 0.3)
    ax.set_title(ttl, fontsize=fontsize)
```
